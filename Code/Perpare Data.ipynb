```
import json
listhmq=[]
qs = json.load(open('data/how_many_qa/HowMany-QA/visual_genome_question_answers.json'))
i=0
for q in qs:
  m=len(q['qas'])
  for i in range(m):
    listhmq.append(q['qas'][i])

for i in range(len(listhmq)):
  listhmq[i]['question_id']=listhmq[i].pop('qa_id')

pathsave="data/how_many_qa/HowMany-QA/qzcreate.json"
json.dump(listhmq, open(pathsave, "w"))
print("The change of VGQA is finished")
```
